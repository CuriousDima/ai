`sentence-transformers`: Specialized for semantic embeddings

```python
from sentence_transformers import SentenceTransformer
from sentence_transformers import util
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
embedding1 = model.encode("I need some food ASAP", convert_to_tensors=True)
embedding2 = model.encode("I am very-very hungry", convert_to_tensors=True)
embedding3 = model.encode("I am angry", convert_to_tensors=True)
```

```python
cos_sims = [util.cos_sim(embedding1, embedding2), util.cos_sim(embedding1, embedding3)]
print(cos_sims)
print(embedding1)
```
